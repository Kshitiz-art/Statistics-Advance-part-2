QUE 4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
sample = [52, 55, 50, 53, 54, 56, 58, 51, 53, 57]  # Replace with your sample data
sample_mean = np.mean(sample)  # Calculate sample mean
sample_std = np.std(sample, ddof=1)  # Calculate sample standard deviation
n = len(sample)  # Sample size

# Population parameters
population_mean = 50  # Specify population mean
alpha = 0.05  # Significance level

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Calculate critical values for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Visualize decision region
z_values = np.linspace(-4, 4, 500)  # Range of Z-values for visualization
pdf_values = norm.pdf(z_values)  # Probability density function of standard normal distribution

plt.figure(figsize=(8, 6))
plt.plot(z_values, pdf_values, label='Standard Normal Distribution')
plt.fill_between(z_values, pdf_values, where=(z_values < -z_critical) | (z_values > z_critical),
                 color='red', alpha=0.4, label='Rejection Region')
plt.axvline(z_statistic, color='blue', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')
plt.axvline(-z_critical, color='green', linestyle='--', label=f'Critical Value (-Z): {-z_critical:.2f}')
plt.axvline(z_critical, color='green', linestyle='--', label=f'Critical Value (+Z): {z_critical:.2f}')
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend(loc='upper right')
plt.grid()
plt.show()

# Calculate P-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Print results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference between sample mean and population mean")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample mean and population mean")
```

QUE 5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0, mu1, sigma, n, alpha):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
        mu0: Population mean under null hypothesis
        mu1: Alternative population mean
        sigma: Standard deviation of the population
        n: Sample size
        alpha: Significance level
    """
    # Calculate the standard error
    std_err = sigma / np.sqrt(n)

    # Critical value for Type 1 error
    z_critical = norm.ppf(1 - alpha)
    critical_value = mu0 + z_critical * std_err

    # Calculate the probability of Type 2 error (Beta)
    beta = norm.cdf(critical_value, loc=mu1, scale=std_err)

    # Visualization
    x = np.linspace(mu0 - 4 * std_err, mu1 + 4 * std_err, 1000)
    null_dist = norm.pdf(x, loc=mu0, scale=std_err)
    alt_dist = norm.pdf(x, loc=mu1, scale=std_err)

    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Hypothesis ($H_0$)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Hypothesis ($H_a$)", color="red")
    plt.axvline(critical_value, color="green", linestyle="--", label="Critical Value")
    plt.fill_between(x, null_dist, where=(x > critical_value), color="blue", alpha=0.3, label="Type 1 Error (Œ±)")
    plt.fill_between(x, alt_dist, where=(x <= critical_value), color="red", alpha=0.3, label="Type 2 Error (Œ≤)")
    plt.title("Type 1 (Œ±) and Type 2 (Œ≤) Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper right")
    plt.grid()
    plt.show()

    # Print errors
    print("Type 1 Error (Œ±):", alpha)
    print("Type 2 Error (Œ≤):", beta)

# Example usage
visualize_errors(mu0=50, mu1=55, sigma=10, n=30, alpha=0.05)
```

QUE 9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(sample, confidence_level):
    """
    Calculates the confidence interval for the sample mean.

    Parameters:
        sample: List or array of sample data
        confidence_level: Confidence level (e.g., 0.95 for 95%)

    Returns:
        Tuple: (lower_bound, upper_bound)
    """
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)

    # Calculate margin of error
    z_critical = norm.ppf((1 + confidence_level) / 2)  # Two-tailed critical Z-value
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [52, 55, 50, 53, 54, 56, 58, 51, 53, 57]
confidence_interval = calculate_confidence_interval(sample_data, 0.95)
print("Confidence Interval (95%):", confidence_interval)
```

QUE 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayesian_inference(prior, likelihood, evidence):
    """
    Calculates posterior probability using Bayes' Theorem.

    Parameters:
        prior: Prior probability of the hypothesis
        likelihood: Likelihood of the evidence given the hypothesis
        evidence: Total probability of the evidence

    Returns:
        Posterior probability
    """
    posterior = (prior * likelihood) / evidence
    return posterior

# Example usage
prior = 0.3  # Prior probability of hypothesis
likelihood = 0.8  # Likelihood of evidence given hypothesis
evidence = 0.5  # Total probability of evidence

posterior = bayesian_inference(prior, likelihood, evidence)
print("Posterior Probability:", posterior)
```

QUE 13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculates expected frequencies for Chi-square test.

    Parameters:
        observed: 2D array of observed frequencies (contingency table)

    Returns:
        Expected frequencies as a 2D array
    """
    row_totals = np.sum(observed, axis=1)
    col_totals = np.sum(observed, axis=0)
    grand_total = np.sum(observed)

    expected = np.outer(row_totals, col_totals) / grand_total
    return expected

# Example usage
observed = np.array([[20, 30], [15, 25], [25, 20]])
expected = calculate_expected_frequencies(observed)
print("Expected Frequencies:")
print(expected)
```

QUE 16  Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

# Generate sample data
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=15, size=30)

# Calculate variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# F-test statistic
f_statistic = var1 / var2

# Degrees of freedom
df1 = len(group1) - 1
df2 = len(group2) - 1

# P-value (two-tailed test)
p_value = 2 * min(f.cdf(f_statistic, df1, df2), 1 - f.cdf(f_statistic, df1, df2))

# Results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: Variances are not significantly different.")
```

QUE 19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
from scipy.stats import shapiro, levene

def check_anova_assumptions(data):
    """
    Checks the assumptions for ANOVA:
    1. Normality
    2. Equal variance
    3. Independence is assumed in the design of the experiment.

    Parameters:
        data: Dictionary of groups and their values.
    """
    # Normality test for each group
    for group, values in data.items():
        stat, p = shapiro(values)
        print(f"Shapiro-Wilk Test for Group {group}: p-value = {p:.4f}")
        if p < 0.05:
            print(f"Group {group} fails the normality test.")
        else:
            print(f"Group {group} passes the normality test.")

    # Equal variance test (Levene's test)
    values_list = list(data.values())
    stat, p = levene(*values_list)
    print(f"\nLevene's Test for Equal Variance: p-value = {p:.4f}")
    if p < 0.05:
        print("Groups fail the equal variance test.")
    else:
        print("Groups pass the equal variance test.")

# Example data
data = {'Group A': [20, 22, 19, 24, 25], 'Group B': [28, 30, 32, 29, 31], 'Group C': [25, 27, 26, 24, 28]}
check_anova_assumptions(data)
```

```python
import pandas as pd
from scipy.stats import f_oneway
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = {'Group': ['A'] * 5 + ['B'] * 5 + ['C'] * 5,
        'Value': [20, 22, 19, 24, 25, 28, 30, 32, 29, 31, 25, 27, 26, 24, 28]}
df = pd.DataFrame(data)

# Perform one-way ANOVA
group_data = [df[df['Group'] == g]['Value'] for g in df['Group'].unique()]
f_statistic, p_value = f_oneway(*group_data)

# Print results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Visualization
sns.boxplot(x='Group', y='Value', data=df)
plt.title("Comparison of Group Means (One-Way ANOVA)")
plt.grid()
plt.show()
```

QUE 23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
from scipy.stats import ttest_1samp

# Simulate random data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, StdDev=5, Size=30

# Perform one-sample t-test
population_mean = 50
t_statistic, p_value = ttest_1samp(sample, population_mean)

# Print results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

QUE 24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
from scipy.stats import chi2

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)
sample_variance = np.var(sample, ddof=1)

# Population parameters
population_variance = 25  # Hypothesized variance

# Chi-square statistic
n = len(sample)
chi_square_statistic = (n - 1) * sample_variance / population_variance

# Calculate p-value (two-tailed test)
p_value = 2 * min(chi2.cdf(chi_square_statistic, df=n-1),
                  1 - chi2.cdf(chi_square_statistic, df=n-1))

# Results
print("Chi-square Statistic:", chi_square_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in population variance.")
```

QUE 25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import statsmodels.stats.proportion as prop

# Example data
success_a, trials_a = 30, 100
success_b, trials_b = 40, 120

# Perform two-proportion Z-test
stat, p_value = prop.proportions_ztest([success_a, success_b], [trials_a, trials_b])

# Results
print("Z-statistic:", stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in proportions.")
```

QUE 26  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import matplotlib.pyplot as plt
from scipy.stats import f

# Generate sample data
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=15, size=30)

# Calculate variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# F-test statistic
f_statistic = var1 / var2

# Degrees of freedom
df1 = len(group1) - 1
df2 = len(group2) - 1

# P-value (two-tailed test)
p_value = 2 * min(f.cdf(f_statistic, df1, df2), 1 - f.cdf(f_statistic, df1, df2))

# Results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Visualization of variances
plt.boxplot([group1, group2], labels=["Group 1", "Group 2"])
plt.title("Comparison of Group Variances")
plt.ylabel("Values")
plt.grid()
plt.show()
```

QUE 27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results

```python
from scipy.stats import chisquare

# Simulated data (observed frequencies)
observed = [50, 30, 20]
expected = [40, 40, 20]  # Expected frequencies

# Perform goodness-of-fit test
chi2_statistic, p_value = chisquare(f_obs=observed, f_exp=expected)

# Results
print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```
